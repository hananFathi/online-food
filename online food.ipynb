## Numerical Feature Analysis

```python
numerical_features = ['Age','Monthly Income','latitude','longitude','Family size']
numerical_features
```

```python
for i in numerical_features:
    plt.figure(figsize=(8,6))
#     if i == 'Family Size':
#         sns.countplot(data=food,x=i,color = 'skyblue')
#     else:
    sns.histplot(food[i],kde = True,color = 'skyblue',bins = 20)
    plt.title(f'Distribution of {i}')
    plt.xlabel(i)
    plt.ylabel('Frequency')
    plt.show()
```

## Categorical Feature Analysis

```python
categorical_feature = ['Gender','Marital Status','Occupation','Educational Qualifications','Feedback']
for i in categorical_feature:
    plt.figure(figsize = (8,4))
    sns.countplot(data=food,x=i,color='skyblue')
    plt.title(f'Distribution of {i}')
    plt.xlabel(i)
    plt.ylabel('Freaquency')
    plt.xticks(rotation = 45)
    plt.show()
```

## Correalation

```python
plt.figure(figsize=(15, 6))
plt.subplot(3,2,1)
sns.boxplot(data=food, x='Age', y='Feedback')
plt.title('Age v/s Feedback')
plt.xlabel('Age')
plt.ylabel('Feedback')

plt.subplot(3,2,2)
sns.boxplot(data=food,x='Age',y='Gender')
plt.title('Age v/s Gender')
plt.xlabel('Age')
plt.ylabel('Gender')
plt.show()
```

```python
numeric_data = food.select_dtypes(include='number')
correlation = numeric_data.corr()

plt.figure(figsize=(10,8))
sns.heatmap(correlation,annot=True,cmap='coolwarm',fmt='.2f',linewidths=0.5)
plt.title('Correlation')
plt.show()
```

## Geographical Analysis

```python
plt.figure(figsize=(10,8))
sns.scatterplot(data = food,x='longitude',y='latitude',alpha =0.5)
plt.title('Geographical Distribution of Customers')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```
